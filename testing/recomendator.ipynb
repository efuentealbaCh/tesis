```python
import pandas as pd
df = pd.read_csv("../data/amazon.csv")
df.head()
```

```python
#Verificamos el tamaño del dataset
df.shape
```

```python
#Verificamos la existencia de posibles datos nulos
print(df.isnull().sum())
#Miramos cuales son los productos que estan con valores nul de acuerdo a la
#exploración realizada anteriormente.
df[df.rating_count.isnull()]
```

```python
#Eliminamos las filas de basandonos en la columna en donde se encuentran los valores nulos.
df.dropna(subset=["rating_count"], inplace=True)
df.isnull().sum()
```

```python
#De una forma similar eliminamos los datos que se encuentran duplicados
df.duplicated().sum()
```

```python
#Se hace un chequeo para el tipo de dato de cada columna del dataFrame
df.dtypes
```

```python
#Algunas variables en un conjunto de datos pueden tener un tipo de datos de objeto, lo que significa que son cadenas.
#Para realizar un análisis numérico de estas variables, necesitamos convertirlas a valores numéricos.

df['discounted_price'] = df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
df['discount_percentage'] = df['discount_percentage'].astype(str).str.replace('%','').astype(float)/100
```

```python
# La columna de calificación tiene un valor con un carácter incorrecto, por lo que excluiremos 
# la fila para obtener un conjunto de datos limpio. 

count = df['rating'].str.contains('\|').sum()
print(f"Total de lineas con '|' na columna 'rating': {count}")
df = df[df['rating'].apply(lambda x: '|' not in str(x))]
count = df['rating'].str.contains('\|').sum()
print(f"Total de linea con '|' na columna 'rating': {count}")
```

```python
df['rating'] = df['rating'].astype(str).str.replace(',','').astype(float)
df['rating_count'] = df['rating_count'].astype(str).str.replace(',','').astype(float)
df.dtypes
```

```python
# Creacion de la columna "rating_weighted"
df['rating_weighted'] = df['rating'] * df['rating_count']

#creacion de columna de sub categorias.
df['sub_category'] = df['category'].astype(str).str.split('|').str[-1]
df['main_category'] = df['category'].astype(str).str.split('|').str[0]
df.columns
```

```python
df.head()
```

```python


#Exploración y visualización de datos
import matplotlib.pyplot as plt
#Analicis de distribuión de productos segun su categoria principal
main_category_count = df['main_category'].value_counts()[:30] #Seleccionamos solo los primeros 30 productos de acuerdo a su categoria principal

plt.bar(range(len(main_category_count)),main_category_count.values)
plt.ylabel('Número de productos')
plt.title('Distribución de categoria principal TOP 30')
plt.xticks(range(len(main_category_count)),'')
plt.show()

top_main_categories = pd.DataFrame({'Main Category': main_category_count.index,'Número de productos':main_category_count.values})
print('Top 30 de categorias principales')
print(top_main_categories.to_string(index=False))
```

```python
sub_category_counts = df['sub_category'].value_counts()[:30]
plt.bar(range(len(sub_category_counts)),sub_category_counts.values)
plt.ylabel('Número de productos')
plt.title('Distribución de sub categorias TOP 30')
plt.xticks(range(len(sub_category_counts)),'')
plt.show()

top_sub_categories = pd.DataFrame({'Sub Category': sub_category_counts.index,'Número de productos':sub_category_counts.values})
print('Top 30 de sub categorias')
print(top_sub_categories.to_string(index=False))
```

```python
#Analisis estadistico de correlación entre las difentes caracateristicas.

numeric_cols = df.select_dtypes(include=['float64','int64'])
correlation_matrix = numeric_cols.corr()
correlation_matrix
```

```python
import seaborn as sns

plt.figure(figsize=(5,5))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
